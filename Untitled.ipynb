```python
import numpy as np
import cv2 
from matplotlib import pyplot as plt
from collections import defaultdict
from sklearn.cluster import OPTICS
```

```python
def plot_img(img, cmap='gray', size=(24,12)):
    plt.figure()
    plt.imshow(img, cmap=cmap)
    plt.axis('off')
    plt.show()
```

```python
img1 = cv2.imread('bbox1.jpg')         
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plot_img(img1)

img2 = cv2.imread('bbox2.jpg')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
plot_img(img2)


img1_2 = cv2.imread('bbox1_2.jpg')         
img1_2 = cv2.cvtColor(img1_2, cv2.COLOR_BGR2RGB)
plot_img(img1_2)

img2_2 = cv2.imread('bbox2_3.jpg')
img2_2 = cv2.cvtColor(img2_2, cv2.COLOR_BGR2RGB)
plot_img(img2_2)

```

```python
def get_cont(img):
    grey = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    mask = cv2.threshold(grey, 255, 255, cv2.THRESH_OTSU)[1]
    img = cv2.bitwise_and(img, img, mask = mask)
    plot_img(img)
    return img

dim = (1, 1)
rectKern = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))

# plot_img(img1)
img1c = get_cont(img1)
# plot_img(img2)
img2c = get_cont(img2)
# plot_img(img1_2)
img1_2c = get_cont(img1_2)
# plot_img(img2_2)
img2_2c = get_cont(img2_2)
```
